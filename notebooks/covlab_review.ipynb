```python
import json
from collections import Counter
import numpy as np


with open("/kaggle/input/convlab/dialogues.json", "r", encoding="utf-8") as f:
    data = json.load(f)


print("ğŸ“¦ Tá»•ng sá»‘ samples:", len(data))
print("ğŸ“‘ Keys cá»§a 1 sample:", data[0].keys())
```

```python
num_turns = []
num_tokens = []

for sample in data:
    turns = sample.get("turns", [])
    num_turns.append(len(turns))

    for t in turns:
        utt = t.get("utterance") or t.get("utterance_en", "")
        num_tokens.append(len(utt.split()))

print("\nğŸ—£ THá»NG KÃŠ Há»˜I THOáº I")
print("â€¢ Trung bÃ¬nh turns / sample:", np.mean(num_turns))
print("â€¢ Min turns:", min(num_turns))
print("â€¢ Max turns:", max(num_turns))
print("â€¢ Tá»•ng utterances:", sum(num_turns))

print("\nğŸ“ Äá»˜ DÃ€I CÃ‚U")
print("â€¢ Trung bÃ¬nh words / utterance:", np.mean(num_tokens))
print("â€¢ Max words:", max(num_tokens))
```

```python
speaker_counter = Counter()

for sample in data:
    for t in sample.get("turns", []):
        speaker_counter[t.get("speaker")] += 1

print("\nğŸ­ Speaker distribution:")
for k, v in speaker_counter.items():
    print(f"  {k}: {v}")
```

```python
emotion_counter = Counter()

for sample in data:
    for t in sample.get("turns", []):
        if "emotion" in t:
            emotion_counter[t["emotion"]] += 1

print("\nğŸ˜Š Emotion distribution:")
for k, v in emotion_counter.items():
    print(f"  {k}: {v}")
```

```python
intent_counter = Counter()

for sample in data:
    for t in sample.get("turns", []):
        acts = t.get("dialogue_acts", {}).get("binary", [])
        for a in acts:
            intent_counter[a.get("intent")] += 1

print("\nğŸ¯ Intent distribution:")
for k, v in intent_counter.items():
    print(f"  {k}: {v}")
```
