## 1. Xem mẫu dữ liệu

```python
import pandas as pd
import os

# Cấu hình hiển thị
pd.set_option('display.max_colwidth', None)

# Đường dẫn file
file_path = r"f:\Projetcs\data_scam\raw\BothBosu\agent_conversation_all.csv"

# Đọc dữ liệu
if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(f"Dataset shape: {df.shape}")
    print("\nCác cột trong file:")
    print(df.columns.tolist())
else:
    print(f"Không tìm thấy file tại: {file_path}")
```
